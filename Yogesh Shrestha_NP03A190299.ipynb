```python
# Importing necessary Library
import pandas as pd #to work with the array
import numpy as np #data science tool for data wrangling and analysis
import matplotlib.pyplot as plt # to visualize data
import seaborn as sns # to make statical graphics
import warnings #to ignire unwanted warnings
warnings.filterwarnings('ignore')
```

```python
def change_to_datetime(df,col): #to change columns object to date time
    df[col] = pd.to_datetime(df[col])
```

```python
for i in['Date_of_Journey','Dep_Time','Arrival_Time']: #using loop to print certain data
    print(i)
    change_to_datetime(data_frame,i)
```

```python
#converting object data type into date time formati
data_frame['journey_day']= data_frame['Date_of_Journey'].dt.day  
data_frame['journey_month']= data_frame['Date_of_Journey'].dt.month
```

```python
data_frame.head()
```

```python
data_frame.drop('Date_of_Journey', axis = 1, inplace = True) #droping unneccessary data colmun
```

```python
data_frame.head(2)
```

```python
#Extra Hour, Minute from Dep_time and Arrival_time and drop Dep_time and Arrival_time
def extract_hour(df, col):
    df[col+'_hour'] = df[col].dt.hour
def extract_minute(df,col):
    df[col+'_minute'] = df[col].dt.minute
def drop_column(df, col):
    df.drop(col,axis = 1, inplace = True)
```

```python
extract_hour(data_frame,'Dep_Time')
extract_minute(data_frame,'Dep_Time')

extract_hour(data_frame,'Arrival_Time')
extract_minute(data_frame,'Arrival_Time')

drop_column(data_frame,'Dep_Time')
drop_column(data_frame,'Arrival_Time')
```

```python
data_frame.head()
```

```python
duration = list(data_frame['Duration']) #list od data inside duration
duration
```

```python
duration[5].split(' ')
```

```python
len(duration[5].split(' '))
```

```python
for i in range(len(duration)): #using for loop 
    if len(duration[i].split(' '))==2:
        pass

    else:
        if 'h' in duration[i]:
            duration[i]= duration[i] + ' 0m'
        else:
            duration[i] = '0h '+duration[i]
```

```python
data_frame['Duration'] = duration #initiallizing data 
data_frame.head()
```

```python
# Separating hour and minute from duration column
duration[5].split(' ')
```

```python
duration[5].split(' ')[1][:-1] #slicling method
```

```python
def hour(x):
    return x.split(' ')[0][:-1]
def minute(x):
    return x.split(' ')[1][:-1]
```

```python
data_frame['Duration_Hour'] =data_frame['Duration'].apply(hour)
data_frame['Duration_Minute'] =data_frame['Duration'].apply(minute)
```

```python
data_frame.head(1)
```

```python
data_frame.drop('Duration', axis = 1,inplace = True)
data_frame.head(1)
```

```python
data_frame.dtypes
```

```python
#converting object types to int type
data_frame['Duration_Hour'] = data_frame['Duration_Hour'].astype(int) 
data_frame['Duration_Minute'] = data_frame['Duration_Minute'].astype(int)
```

```python
data_frame.dtypes
```

```python
categorical_column = [col for col in data_frame.columns if data_frame[col].dtype=='O'] # to Fetch Object Column
```

```python
# Handling Route Column
```

```python
categorical_data['Route_01'] = categorical_data['Route'].str.split('→').str[0]

categorical_data['Route_02'] = categorical_data['Route'].str.split('→').str[1]
categorical_data['Route_03'] = categorical_data['Route'].str.split('→').str[2]
categorical_data['Route_04'] = categorical_data['Route'].str.split('→').str[3]
categorical_data['Route_05'] = categorical_data['Route'].str.split('→').str[4]
categorical_data['Route_06'] = categorical_data['Route'].str.split('→').str[5]
```

```python
categorical_data.drop('Route', axis =1 , inplace = True)
```
